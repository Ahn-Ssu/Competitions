```python
from rdkit.Chem import MolFromSmiles
featurizer = dc.feat.MolGanFeaturizer(max_atom_count=30)

rdkit_mol = MolFromSmiles(smiles[0])
features = featurizer([rdkit_mol, smiles[0]])

print(features.shape)
print(features[0])
```
